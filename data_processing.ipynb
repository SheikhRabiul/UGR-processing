```python
import os
import csv
startdir='./UGRdata/raw/'
#outputdir = './UGRdata/parsed/'
suffix='.csv'
#'2016-03-18 00:00:00'
# 01234567890123
hour_index = 13
date_index = 11
#hour_range = range(1, )

header =['DateTime','FlowDuration','SrcIP','DstIP','SrcPort','DstPort','Protocol','Flag','ForwardingStatus','TypeofService','PacketExed','Bytes','TrafficType','Label']

threshold = 24
    
for root, dirs, files in os.walk(startdir):
    for name in files:
        if name.endswith(suffix):
            filename=os.path.join(root,name)
            
            with open(filename) as fd:        # open the csv file
                rd = csv.reader(fd)           #  as a csv input file
                old = None                    # no previous line
                i = 0                         # we will start numbering output files with 1
                #header = next(rd)             # store the header line
                fdout = None
                wr = None
                counter = 0
   
                print('file: ',filename)
                for row in rd:
                    #print('row: ',row)
                    if len(row)==0 or 'DateTime' in row[0]:
                        #print("row 0, i=", i)
                        continue
                   
                    #_min = int(row[0][hour_index+1:hour_index+3])
                    #_dt = row[0][:hour_index] + "_"+ str(int(_min/10)*10).zfill(2)
                    #by 2 hour
                    #_hour = int(row[0][date_index:hour_index])     
                    #_dt = row[0][:date_index] + str(int(_hour/2)*2).zfill(2)+ '_00'  
                    
                    #by 1 hour 
                    _dt = row[0][:hour_index] + "_00"
                    
                    if _dt != old:     # we have a different hour (or the first one...)
                        old = _dt      # store current time for next rows
                        print('new time:', old)
                        i += 1                # increase output file number
                        if i > threshold:
                            break
                        if fdout is not None:   # eventually close previous output file
                            fdout.close()
                            print("Time Index", i-1, ": ", counter)
                        out_file = (filename[:-4] + '_' + old + filename[-4:]).replace('/raw/','/split_1h/')
                        fdout = open(out_file, 'w', newline='')     # open a new output file
                        wr = csv.writer(fdout, quoting=csv.QUOTE_ALL)  # with expected csv params
                        _ = wr.writerow(header)   # write the header
                        counter = 1
                    else:
                        if "background" in row[-1]:
                            row.append('normal')
                        elif "blacklist" in row[-1] and 'blacklist' in filename:
                            row.append('abnormal')
                        elif "spam" in row[-1] and 'spam' in filename:
                            row.append('abnormal')
                        elif "sshscan" in row[-1] and 'sshscan' in filename:
                            row.append('abnormal')
                        else:
                            row.append('abnormal')

                        _ = wr.writerow(row)      # copy the row to the current output file
                        counter += 1
                
                if fdout is not None:
                    fdout.close()
```

# 2. Aggregate

```python
import os
import csv
from datetime import datetime
import pandas as pd
import sys

startdir='./UGRdata/split_1h/'
_dir = '/split_1h/'
suffix='.csv'


threshold = sys.maxsize


#dic
# key: label, srcip, dstip, srcport, dstport, protocol
# value:
# 0 counter
# 1 sum_flowduration = 0
# 2 sum_forwarding = 0
# 3  sum_typeservice = 0
# 4 sum_packageexed = 0
# 5 sum_bytes = 0


    
for root, dirs, files in os.walk(startdir):
    for name in files:
        if name.endswith(suffix):
            filename=os.path.join(root,name)
            dic={}
            with open(filename) as fd:        # open the csv file
                #march.week3_2016-03-18 10_56.csv
                #            9876543210987654

                #if "march" in filename:
                #   continue
                
                dt = (filename[-20:-4]+":00").replace('_',':')
                rd = csv.reader(fd)           #  as a csv input file
                old = None                    # no previous line
                i = 0                         # we will start numbering output files with 1
                header = next(rd)             # store the header line
                fdout = None
                wr = None
                counter = 0

                print('file: ',filename)
                print('file size: ',os.path.getsize(filename))

                sum_abnormal = 0
                for row in rd:
                    if len(row) != 14 or 'SrcIP' in row:
                        #print('header')
                        continue

                    #Label, SrcIP,DstIP,SrcPort,DstPort,Protocol
                    _key = row[13] + ','+ row[2] + ','+ row[3]+','+row[4]+','+row[5]+','+row[6]
                    #['DateTime','FlowDuration','SrcIP','DstIP','SrcPort',
                    #'DstPort','Protocol','Flag','ForwardingStatus','TypeofService',
                    #'PacketExed','Bytes','TrafficType','Label']
                    if _key in dic:
                        dic[_key][1] += float(row[1])
                        dic[_key][2] += float(row[8])
                        dic[_key][3] += (dic[_key][3]*dic[_key][0] + float(row[9]))/(dic[_key][0] + 1)
                        dic[_key][4] += float(row[10])
                        dic[_key][5] += float(row[11])
                        dic[_key][0] += 1

                    else:
                        # 0 counter 1 sum_flowduration = 0 2 sum_forwarding = 0 
                        # 3  sum_typeservice = 0 4 sum_packageexed = 0  5 sum_bytes = 0
                        new_value = [1,0,0,0,0,0]
                        dic[_key] = new_value
                
                #output dic to file
                out_file = (filename).replace(_dir,'/aggr/')
                with open(out_file, 'w') as f:
                    for k in dic.keys():
                        #DateTime, key..., value...
                        f.write(dt+','+k + ','+','.join(str(x) for x in dic[k]) +'\n')
                print(out_file)
                print("writen done! file size: ", os.path.getsize(out_file))

            
```

```python
import os
import csv
from datetime import datetime
import pandas as pd
import sys
import socket, struct


startdir='./UGRdata/aggr/'
_dir = '/aggr/'
suffix='.csv'


start_dt = '2016-03-18 10:00:00'
end_dt = '2016-03-19 15:00:00'

init_dt_str = '2016-03-18 00:00:00'

def convert_int_ip(ipint):
    #ipint=16909060
    return socket.inet_ntoa(struct.pack('!L', ipint))
    
def convert_ip(ipstr):
    if ipstr.strip == "":
        return 0
    return struct.unpack("!L", socket.inet_aton(ipstr.strip()))[0]

def convert_prot(pstr):
    switcher = {
        "TCP": 1,
        "UDP": 2,
        "ICMP": 3,
        "ESP": 4,
        "IPIP": 5,
        "GRE": 6,
    }
    return (switcher.get(pstr, 0))

def get_header():
    return ['Label','Timestamp', 'FlowDuration', 'SrcIP', 'DstIP', 'SrcPort','DstPort','Protocol',
            'ForwardingStatus','TypeofService','PacketExed', 'Bytes']

#Input
#DateTime, key..., value...
#0:DateTime 1:Label 2:SrcIP 3:DstIP 4:SrcPort 5:DstPort 6:Protocol 
#7:Counter 8:Sum_FlowDuration 9:Sum_Forwarding 10:Sum_Typeofserver 11: Sum_packageexchanged 12:Sum_bytes 

def get_newrow(row):
    new_row = []
    dt_str = row[0].strip()
    lab = row[1].strip()
    srcip = row[2].strip()
    dstip = row[3].strip()
    srcport = row[4].strip()
    dstport = row[5].strip()
    prot = row[6].strip()
    num_conn = row[7].strip()
    sum_flowduration = row[8].strip()
    sum_forwarding = row[9].strip()
    sum_typeofservice = row[10].strip()
    sum_packageexed = row[11].strip()
    sum_bytes = row[12].strip()
    
    #label
    new_row.append(lab)
    #Timestamp
    dt_obj = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    diff = dt_obj - init_dt
    diff_sec = diff.seconds
    new_row.append(diff_sec)
    #FlowDuration
    new_row.append(sum_flowduration)
    #SrcIP
    new_row.append(convert_ip(srcip))
    #DstIP
    new_row.append(convert_ip(dstip))
    #SrcPort
    new_row.append(srcport)
    #DstIP
    new_row.append(dstport)
    #Protocol
    new_row.append(convert_prot(prot))
    #ForwardingStatus
    new_row.append(sum_forwarding)
    #'TypeofService',
    new_row.append(sum_typeofservice)
    #'PacketExed', 
    new_row.append(sum_packageexed)
    #'Bytes'
    new_row.append(sum_bytes)
    return(new_row)
    

counter_n = 0
counter_ab = 0



    
for root, dirs, files in os.walk(startdir):
    init_dt = datetime.strptime(init_dt_str, '%Y-%m-%d %H:%M:%S')
    for name in files:
        if name.endswith(suffix):
            filename=os.path.join(root,name)

            with open(filename) as fd:        # open the csv file
                dt = (filename[-20:-4]+":00").replace('_',':')
                if dt < start_dt or dt >= end_dt:
                    continue

                rd = csv.reader(fd)           # as a csv input file
                print('selected: ',filename)

                output_filename = filename.replace(_dir,'/converted/')

                with open(output_filename, 'w') as outfile:
                    csv_writer = csv.writer(outfile, delimiter=',')
                    csv_writer.writerow(get_header())

                    for row in rd:
                        if len(row) == 0 or 'SrcIP' in row:
                            print('header')
                            continue

                        if 'abnormal' in row[1]:
                            counter_ab += 1
                        else:
                            counter_n += 1

                        csv_writer.writerow(get_newrow(row))

print('# of normal: ',counter_n)
print('# or abnormal: ',counter_ab)
    
```
